```python
from sklearn import datasets
import numpy as np

# Load the Iris dataset
iris_data, iris_labels = datasets.load_iris(return_X_y=True, as_frame=False)

# Choose a continuous feature (e.g., sepal length, index 0)
feature_index = 0
selected_feature = iris_data[:, feature_index]

# Compute overall statistics
mean_all = np.mean(selected_feature)
min_all = np.min(selected_feature)
max_all = np.max(selected_feature)
variance_all = np.var(selected_feature)
sd_all = np.std(selected_feature)
instance_count_all = len(selected_feature)

print(f"Overall Statistics for Feature {feature_index}:")
print(f"  Instance Count: {instance_count_all}")
print(f"  Mean: {mean_all}")
print(f"  Min: {min_all}")
print(f"  Max: {max_all}")
print(f"  Variance: {variance_all}")
print(f"  Standard Deviation: {sd_all}")

# Split the dataset by class labels
class_0_feature = selected_feature[iris_labels == 0]
class_1_feature = selected_feature[iris_labels == 1]
class_2_feature = selected_feature[iris_labels == 2]

# Compute statistics for each class
def compute_stats(feature_array, class_label):
    mean_class = np.mean(feature_array)
    min_class = np.min(feature_array)
    max_class = np.max(feature_array)
    variance_class = np.var(feature_array)
    sd_class = np.std(feature_array)
    instance_count = len(feature_array)

    print(f"\nStatistics for Class {class_label}:")
    print(f"  Instance Count: {instance_count}")
    print(f"  Mean: {mean_class}")
    print(f"  Min: {min_class}")
    print(f"  Max: {max_class}")
    print(f"  Variance: {variance_class}")
    print(f"  Standard Deviation: {sd_class}")

compute_stats(class_0_feature, 0)
compute_stats(class_1_feature, 1)
compute_stats(class_2_feature, 2)
```

# Section 3 - Split into training and test sets

```python
import numpy

# Randomly split the data into 50:50 training:test sets
random_seed = 1
rng =  numpy.random.default_rng(random_seed)
rand_inds = numpy.arange(iris_labels.shape[0],)
rng.shuffle(rand_inds)
split_point = int(0.5*iris_labels.shape[0])

training_data = iris_data[rand_inds[0:split_point], :]
training_labels = iris_labels[rand_inds[0:split_point]]
test_data = iris_data[rand_inds[split_point:iris_labels.shape[0]], :]
test_labels = iris_labels[rand_inds[split_point:iris_labels.shape[0]]]

print("Size of the training data:", training_data.shape)
print("Size of the ttest data:", test_data.shape)
```

# Section 4 - Train and evaluate a kNN model

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

model_kNN = KNeighborsClassifier(n_neighbors=5)
model_kNN.fit(training_data, training_labels)
test_predictions_kNN = model_kNN.predict(test_data)

print("\n What proportion of the kNN test predictions were correct? %.2f " % accuracy_score(test_labels, test_predictions_kNN))
```

##### Metrics

```python
import numpy as np
from sklearn.metrics import confusion_matrix

def calculate_metrics(actual_labels, predicted_labels):
    """Calculates metrics using NumPy."""

    # Overall Accuracy
    accuracy = np.mean(actual_labels == predicted_labels)
    print(f"Accuracy: {accuracy:.4f}")

    # Confusion Matrix
    conf_matrix = confusion_matrix(actual_labels, predicted_labels)
    print("\nConfusion Matrix:")
    print(conf_matrix)

    # Metrics per class
    unique_labels = np.unique(actual_labels)

    for label in unique_labels:
        # True Positives, False Positives, False Negatives
        tp = conf_matrix[label, label]
        fp = np.sum(conf_matrix[:, label]) - tp
        fn = np.sum(conf_matrix[label, :]) - tp

        # Precision, Recall, F1-score
        if tp + fp == 0:
          precision = 0
        else:
          precision = tp / (tp + fp)

        if tp + fn == 0:
          recall = 0
        else:
          recall = tp / (tp + fn)

        if precision + recall == 0:
          f1 = 0
        else:
          f1 = 2 * (precision * recall) / (precision + recall)

        print(f"\nClass {label}:")
        print(f"  Precision: {precision:.4f}")
        print(f"  Recall: {recall:.4f}")
        print(f"  F1-score: {f1:.4f}")

calculate_metrics(test_labels, test_predictions_kNN)
```

#### Visuals

##### Confusion Matrix

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

def plot_confusion_matrix(actual_labels, predicted_labels, classes):
    """Plots a confusion matrix using matplotlib."""

    conf_matrix = confusion_matrix(actual_labels, predicted_labels)
    print(conf_matrix)

    plt.figure(figsize=(8, 6))
    plt.imshow(conf_matrix, interpolation='nearest', cmap=plt.cm.Blues)
    plt.title('Confusion Matrix')
    plt.colorbar()

    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    thresh = conf_matrix.max() / 2.
    print(thresh)
    print(conf_matrix.shape)
    print(np.ndindex(conf_matrix.shape))
    
    for i, j in np.ndindex(conf_matrix.shape):
        plt.text(j, i, format(conf_matrix[i, j], 'd'),
                 ha="center", va="center",
                 color="white" if conf_matrix[i, j] > thresh else "black") # for the text vs background colour

    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.tight_layout()
    plt.show()

# Example usage (replace with your actual data)
class_names = ['Class 00', 'Class 1', 'Class 2']  # Replace with your class names

plot_confusion_matrix(test_labels, test_predictions_kNN, class_names)
```

##### Scatter Plots 

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_feature_pairs_with_errors(data, labels, predictions, feature_names, class_names):
    """Plots scatter plots with errors highlighted."""

    num_features = data.shape[1]
    num_classes = len(np.unique(labels))

    feature_pairs = [(i, j) for i in range(num_features) for j in range(i + 1, num_features)]

    # Explicitly assign colors to classes
    col = ['blue', 'red', 'green']

    for pair in feature_pairs:
        feature1_index, feature2_index = pair
        feature1_name = feature_names[feature1_index]
        feature2_name = feature_names[feature2_index]

        plt.figure(figsize=(8, 6))

        # Plot all data points
        for class_label in range(num_classes):
            class_data = data[labels == class_label]
            plt.scatter(
                class_data[:, feature1_index], 
                class_data[:, feature2_index], 
                label=class_names[class_label],
                color = col[class_label]
            )

        # # Highlight incorrect predictions
        incorrect_indices = np.where(labels != predictions)[0]

        incorrect_data = data[incorrect_indices]
        incorrect_labels = labels[incorrect_indices]
        incorrect_predictions = predictions[incorrect_indices]

        print(incorrect_data, incorrect_labels, incorrect_predictions)

        for class_label in range(num_classes):
            class_data = incorrect_data[incorrect_labels == class_label]
            
            plt.scatter(
                class_data[:, feature1_index], 
                class_data[:, feature2_index], 
                color = 'black',
                marker='x'
            )

        plt.xlabel(feature1_name)
        plt.ylabel(feature2_name)
        plt.title(f"Scatter Plot: {feature1_name} vs {feature2_name}")
        plt.legend()
        plt.show()


# Example usage (replace with your actual data)
# Assuming you have:
# test_data, test_labels, test_predictions_kNN, feature_names, class_names
# from your notebook

plot_feature_pairs_with_errors(test_data, test_labels, test_predictions_kNN, iris_bunch.feature_names, class_names)
```

# Section 6 - Explore the effect of the kNN hyperparameters

* Try different values of k, i.e. the number of nearest neighbours, e.g. k = 1, 2, 5, 10, 20. What effect of k do you notice?
* Try a different distance metric. See https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsClassifier.html.

```python
%matplotlib inline

import matplotlib.pyplot as plt

# Set up different values of k
ks = [1, 2, 5, 10, 20]

# Initialize variables
acc_euclidean = numpy.zeros(len(ks),)
acc_manhattan = numpy.zeros(len(ks),)
counter = 0


for k in ks:

  # Train and evaluate kNN model with Euclidean distance metric
  model_kNN_euclidean = KNeighborsClassifier(n_neighbors=k, metric='euclidean')
  model_kNN_euclidean.fit(training_data, training_labels)
  acc_euclidean[counter] = accuracy_score(test_labels, model_kNN_euclidean.predict(test_data))



  # Train and evaluate kNN model with Manhattan distance metric
  model_kNN_manhattan = KNeighborsClassifier(n_neighbors=k, metric='manhattan')
  model_kNN_manhattan.fit(training_data, training_labels)
  acc_manhattan[counter] = accuracy_score(test_labels, model_kNN_manhattan.predict(test_data))

  counter += 1



# Plot data
plt.plot(ks, acc_euclidean, "b.-", linewidth=1, label='Euclidean')
plt.plot(ks, acc_manhattan, "k*--", linewidth=1, label='Manhattan')

plt.xlabel('k', size=20)
plt.xticks(numpy.arange(0, 21, step=5))
plt.ylabel('Accuracy', size=20)
plt.title('Comparing different values for kNN hyperparameters', size=20)
plt.legend(title='Distance metrics:')

plt.show()
```

# Section 7 - Train and evaluate a Bagging model

```python
from sklearn.ensemble import BaggingClassifier
import math

# Set the max number of features to be used to split each node for each tree
max_feats = int(math.sqrt(training_data.shape[1]))

model_B = BaggingClassifier(n_estimators=10, max_features=max_feats, random_state=random_seed)
model_B.fit(training_data, training_labels)
test_predictions_B = model_B.predict(test_data)


print("\n What proportion of the Bagging test predictions were correct? %.2f " % accuracy_score(test_labels, test_predictions_B))
```
