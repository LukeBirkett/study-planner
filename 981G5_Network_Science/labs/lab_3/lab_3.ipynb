### Brief

> On a construction site, some stakes have been hammered into the ground. As you watch, a worker ties a rope to one of the stakes and then runs it around another stake, some distance away, pulling it tight. When you ask the worker what they are doing, they say, ‘See, I need to run a cable from stake A over there, via stake B, here, to stake C all the way over there. So I tie this rope to A, wind it around B and then tie the end to C. This way I know how long a cable I need to buy.’ 
> 
> This makes you wonder: When is it possible to measure the sum of the pairwise distances between all pairs of stakes, by using a single rope in this way?
> 

---

### Answer and Notes

This isn't a problem involving complex situtations. Instead it requires you to think about **Graph Topology** and **Path Theory**. 

The key area for this puzzle is **Eulerian Paths**. The goal is to trace out a path without "lifting the pencil" and covering each edge only exactly once. Another way to view this latter term is that once an edge is passed, it is turned off. 

The question is asking us to **specific the conditions** from which this goal is possible to achieve. 

There are only really two main parameters to look at **Nodes** and **Edges**. But we can combine the two by looking at **Node Degrees**. 

In the end, there should be a finding that revolves around how many "odd-degree" nodes a network can have if you want to cover it in one clean circuit. 

Within this question, it is important to think about **"Complete Graphs"** ($K_n$). We are asked about the **"sum of all pairwise distances between all pairs of stakes."** If a graph is complete, then there is a pair (edge) between every single node which the circuit needs to cover exactly onces. 

The worker’s goal is not just to connect the stakes (which a tree would do); his goal is to measure every single **pairwise distance** (the **Complete Graph**). This means that touching every node is simply not good enough, you needs to traverse every edge in the complete graph. If the graph was just a tree, the worker's logic ($A \to B \to C$) would work perfectly. Because it's a complete graph, it fails for almost all $N$.

There are few different things to know, or workout, in order to know if this problem can be acheived:
* How many edges are there in a "Complete Graph" of $N$ stakes?
* Can a single path (the rope) cover every one of those edges without overlapping?
* For what value of $N$ (number of stakes) does the math actually allow an Eulerian Path?

The question states **"Pairwise distances between all pairs"**. This seems like a complicated phrase but pairwise is just a way to respresent as edge between two specific nodes. For example if you have 3 nodes $(A, B, C)$ the pairs are $AB$, $AC$, and $BC$ and the "sum" is $AB + AC + BC$.

Based on this small example where $N=3$, if we were to use a rope of length: $AB + BC$, then the rope does not measure the sum of all pairwise distances because it is missing $AC$. Though if we use a rope of length: $AB + BC + AC$ then it does but on the condition that it ends up where it started. 

Remeber, the rules are that the "rope" must be continuous, cover evey possible edge and cannot traverse the same edge twice. There isn't anything strictly that forbids it from ending up at the begining point but it will be interesting to see if is the only possible outcome. Note, that one specific thing to highlight is that if a route coverers $AB + BC$ then it touches all nodes but not all edges.

For $N=3$, we have established that we can use a single rope that measures the sum of all pairwise distances but it will end up at the start point. In graph theory this is known as a **circuit**. If we end up at a different point than we started then we have a **Path**.

A graph of $N=4$ has **6 pairwise distances** to measure (the 4 sides of a square plus the 2 diagonals). Additionally all four nodes have **degrees of 3**.

You cannot trace a four nodes graph if just one clean pass. This is known as the **Eulerian Path rule**. 

To trace a graph with one rope without overlapping, you look at the degree (number of connections) of each node:
* A graph can only be traced in one go if it has **zero or exactly two nodes** with an **odd degree**.
* This is because everytime you enter and leave a node, you use 2 connections. This means to satisfy all nodes, these ones have to be the start and/or end, otherwise, completing one if the middle will end the route as you will be trapped. 

The question asks "When is it possible...?" You are looking for the values of $N$ where a Complete Graph ($K_n$) has an Eulerian Path.

In a complete graph with $N$ nodes, every node has the same degree of $d = N - 1$.

Remeber we were able to get our $N=3$ example to work. Here, there were $3$ nodes where each had a degree of $2$ (Even). Does this mean even is an alternative requirement for this problem to be solved? (And therefore if $d = N-1$ then N needs to be odd)

---

### Two Conditions for Solvability

1. Have exactly $2$ odd degree nodes.
2. Have only even degree nodes. 

> * If $N=2$, the degree is 1 (2 odd nodes — Works!)
> * If $N=3$, the degree is 2 (0 odd nodes — Works!)
> * If $N=4$, the degree is 3 (3 odd nodes — Doesn't Work)
> * If $N=5$, the degree is 4 (0 odd nodes — Works!)

With $N=5$ every node has a degree of $4$. This is important because it means every time the we enter a node we have an an **exit path** available that hasn't been used yet.

However, one thing to note is that with $N=5$ we again find ourselve in a situation where we end up at the starting point creating a **circuit**. It is only when a graph has exactly **two odd-degree nodes** that where you can great a **path**. With a complete graph this is only possible is $N=2$ so it is a bit of a secret rule to allow for this particular case. 

This means the conditions where is it possible to "measure the sum of the pairwise distances between all pairs of stakes, by using a single rope" is: 
* **If $N=2$:** The worker can tie to A and C. 
* **If $N=$ any odd number (3, 5, 7...):** It is possible to measure all distances, but the worker would have to tie both ends of the rope to the same stake (A to A), creating a **circuit**.

> **If $N$ is even ($N > 2$):** It is impossible.

---

### Does this carry over to non-complete graphs?

Yes, the Eulerian rules are universal laws of graph topology meaning the odd rule holds true.
* If there are 0 odd nodes (**i.e. all even nodes**) you can complete a **circuit** but you must end up where you began. 
* If there are **2 odd nodes** you must begin at one and end at the other creating a path. 
* If there are **more than 2 odd nodes**, it is physically impossible to cover every edge exactly once with one rope, regardless of how the nodes are positioned.

### Scale-Free Network: Bitcoin/Finance/Lending

Financial Networks, particularly modern tech-based ones, tend to mimic scale-free networks. For example, with respect to credit, there tends to be a naturally **flocking to reputable entities**. New people/entities looking for credit are more likely to connect with (and rate) these established hubs rather than a random node, i.e. just a person down the pub. This is a classic example of **Preferential Attachment**. Due to this, you will likely see a very high-degree "core" in the $k$-shell analysis of these networks. When you "attack" the network by removing high-degree hubs, the **Largest Connected Component (LCC)** will likely plummet rapidly, showing that the "credit-worthiness" of the entire system relies on a few key players. If you were modeling systemic risk in a banking network, you'd treat it as a scale-free network. You’d want to know: "If Lehman Brothers (a hub) fails, does the whole LCC collapse?"

In a FinTech ecosystem, a provider like Stripe (payments), Plaid (data aggregation), or AWS (infrastructure) acts as a **high-degree hub**. Hundreds of smaller "Neo-banks" or apps connect to these few hubs. This creates a scale-free structure that is very efficient but has a single point of failure: if Plaid's API goes down, thousands of independent apps lose their ability to function. An outage at one identity-verification API can "de-link" hundreds of fintechs from their customers, effectively "shattering" the connected market

## Scale-Free: [Bitcoin OTC](http://konect.cc/networks/soc-sign-bitcoinotc/)

The Bitcoin OTC (Over-The-Counter) dataset is one of KONECT’s most famous "Web of Trust" networks, representing a who-trusts-whom system among users of the Bitcoin OTC marketplace. Because Bitcoin users are pseudonymous, this platform was created to allow traders to rate one another to mitigate the risk of fraud in peer-to-peer transactions.

> Source: S. Kumar, F. Spezzano, V.S. Subrahmanian, C. Faloutsos. [Edge Weight Prediction in Weighted Signed Networks](https://cs.stanford.edu/~srijan/wsn/). IEEE International Conference on Data Mining (ICDM), 2016.

### Dataset Overview
* **Nodes:** 5,881 users (traders).
* **Edges:** 35,592 ratings.
* **Network Type:** Directed, Weighted, and Signed.
* **Domain:** Social / Trust & Financial.

---

### Key Structural Features

* **Signed Weighted Edges:** Unlike simple "friend" networks, each edge carries a score from -10 (total distrust) to +10 (total trust). This makes it a rare example of a "signed" network where negative ties (distrust) are explicitly recorded. Approximately 89% of the ratings are positive.
* **The "Reputation Hub" Effect:** The network exhibits a strong scale-free degree distribution. In a credit-like environment, successful traders act as "hubs" with thousands of incoming positive ratings, while risky or new users have very few.
* **Temporal Nature:** Each rating is timestamped, allowing researchers to study how reputation evolves over time or how a "trust cascade" might form when a previously trusted hub is suddenly flagged as fraudulent.

> Note, the "sign" is whether the value is positive or negative and the "weight" is value itself. 

Because it relies on a few highly reputable hubs to maintain the "integrity" of the marketplace, it is structurally robust to the random removal of users but extremely fragile to a targeted attack. If you remove the top 5% of the most-trusted hubs, the "connected component" of trust will likely shatter, as those hubs are the only bridges connecting different sub-communities of traders.

The resilience of the Bitcoin OTC network isn't just about connectivity; it’s about Opinion Dynamics. If the 'hubs' are removed, the 'Voter Model' of trust fails because there are no longer enough bridges to facilitate a consensus on who is a 'good' trader.

> Note, in a Directed graph, there is a difference between a "Weakly" and "Strongly" Connected Component. For this lab, using the Weakly Connected Component (treating edges as bidirectional) is usually the standard way to visualize fragmentation.

---

#### The Flexibilty and Utility of an Directed, Signed and Weighted Network

> The inherent complexity of a signed, directed, and weighted dataset like Bitcoin OTC provides a unique analytical advantage, as it essentially exists as a "superposition" of multiple network layers that can be isolated through strategic simplification. By treating the data as a simple undirected graph, researchers can diagnose the raw **topological robustness** of the marketplace, identifying the fundamental physical connectivity that persists regardless of sentiment. Alternatively, by filtering for positive edge weights, the network is transformed into a **functional trust web**, revealing the "credit core" of reputable entities that sustain the ecosystem’s integrity. Retaining the directionality further allows for the study of **hierarchical influence**, distinguishing between nodes that act as "authorities" (high in-degree) and those that act as "auditors" (high out-degree). This flexibility ensures that a single dataset can be "sliced" to isolate specific variables—such as systemic risk, reputational hierarchy, or adversarial conflict—without losing the underlying context of the complex system.
> 
> The Bitcoin OTC network serves as an ideal substrate for studying opinion dynamics and voter models because its signed architecture allows for the simulation of both imitation and reactance. In a standard voter model, nodes typically adopt the opinions of their neighbors; however, the inclusion of negative trust ratings enables the modeling of social polarization, where a "distrust" edge might trigger a node to adopt the diametrically opposite opinion of its neighbor. Furthermore, the weighted edges provide a mechanism for "persuasion strength," where a $+10$ rating exerts significantly more influence on an agent's state than a $+1$ rating. When combined with the network's temporal timestamps, these features allow researchers to track "trust cascades" in real-time, observing how the market reaches a consensus on a trader's reputation or how "adversarial hubs" can effectively quarantine misinformation by preventing its spread through hostile links.

**Points to highlight:**
1. **The "p-value" trap:** In the KONECT handbook (and the Clauset et al. methodology they use), a high p-value (usually > 0.1) means the Power Law is a plausible hypothesis. Your p-values are 0.00, which mathematically tells a researcher: "This network looks scale-free to the eye, but there are significant deviations from a perfect mathematical power law." This is very common in real-world human networks.
2. **In-degree vs. Out-degree:** Note that $γ_{3,i}$ (2.27) is higher than $γ_{3,o}$ (2.06).

    - In-degree is "reputation" (how many ratings you receive).
    - The higher exponent for In-degree suggests that it is slightly "harder" to become a massive trust-hub than it is to simply be a very active rater.

These metrics focus only on the "Tail" (the most connected nodes). They ignore the thousands of users with only 1 or 2 ratings and focus on the "whales" or hubs. Even if the p-value is low (suggesting the fit isn't perfect), the exponents ($\gamma$) being around 2.0–2.3 confirms that you have a "Heavy-Tailed" distribution, which is the functional requirement for the scale-free behavior you'll be testing in your lab.

#### Is Bitcoin OTC Actually Scale-Free

The classification of the Bitcoin OTC network as "scale-free" reveals a fascinating tension between mathematical theory and functional reality. While the extreme disparity between the **maximum degree (1,298)** and the **average degree (12.1)** points to a clear hub-and-spoke architecture, the statistical **$p$-values of 0.00** technically reject a "perfect" power law fit. This discrepancy often occurs in real-world systems where finite-size effects — such as the limited number of traders — prevent the distribution from reaching the infinite mathematical ideal. However, with a power-law exponent of $\gamma \approx 1.97$, the network falls into the "ultra-scale-free" regime. In this state, the hubs are so dominant that they essentially "own" the connectivity of the system, acting as critical reputation anchors. For the purposes of network robustness, the system behaves exactly like a scale-free model: it is remarkably resilient to random failures but exceptionally fragile to targeted attacks on its primary "super-hubs."

> "Visually, the Bitcoin OTC degree distribution appears as a Power Law with an Exponential Cutoff. While the initial scaling regime ($10^0$ to $10^1$) is linear, the distribution 'drifts' into a shallower slope as it approaches the highest degrees ($10^2$–$10^3$). This suggests that while the network is governed by a 'rich-get-richer' mechanism, the 'super-hubs' eventually encounter physical or temporal constraints that prevent a perfect power-law fit. Despite this 'bending,' the network remains functionally scale-free because its connectivity spans three orders of magnitude, a feature entirely absent in grid-like or random topologies."

##### Data Considerations:
- **Directed Network:** The data file has to and from node columns confirming that this is a directed network. - **Signs/Weights and k-Clustering:** Ratings range from -10 to +10. For K-Core, NetworkX ignores weight and only looks at connectivity. It might be a good sign to also only look at **positive** connections as this will allow us to analyse a "trust network".

### Initalising the Network Using NetworkX

```python
import networkx as nx

# Define the file path
file_path = './soc-sign-bitcoinotc/out.soc-sign-bitcoinotc'

# Load the graph
# We skip comments (lines starting with %) and specify the column types
G = nx.read_edgelist(
    file_path,
    comments='%',
    nodetype=int,
    data=[('weight', int), ('timestamp', float)],
    create_using=nx.DiGraph() # Use Directed Graph as per metadata
)

print(f"Nodes: {G.number_of_nodes()}, Edges: {G.number_of_edges()}")
```

#### Initalising using Pandas

If the `.out` file has extra metadata lines at the top (often starting with `%`), Pandas is slightly better at skipping them and handling the whitespace.

```python
import pandas as pd
import networkx as nx

# Load with Pandas, skipping the metadata/header lines (usually starting with %)
# 'sep' specifies any whitespace as the delimiter
df = pd.read_csv('./soc-sign-bitcoinotc/out.soc-sign-bitcoinotc', sep=r'\s+', comment='%', header=None, 
                 names=['source', 'target', 'weight', 'timestamp'])

# Convert to NetworkX
pG = nx.from_pandas_edgelist(df, 'source', 'target', edge_attr=['weight', 'timestamp'], 
                             create_using=nx.DiGraph())

print(f"Nodes: {pG.number_of_nodes()}, Edges: {pG.number_of_edges()}")
```

## K-Core

### The $k$-Core Decomposition: A Summary

The $k$-**core** of a graph is the maximal subgraph where every participating node is connected to at least $k$ other nodes within that same subgraph. Unlike simple degree filtering, which only looks at a node's local connectivity, $k$-core decomposition is an iterative process that strips away the "periphery" of a network to reveal its **structural backbone**.

> A **0-core** is the entire graph (every node has at least zero edges). The 1-core consists of all nodes that have at least one connection after isolated nodes are removed. This is not just simply every node with an edge. That edge has to connected to another node which is also connected to the group of connected nodes, i.e. 2 nodes completely seperated from any other nodes each have 1 edge but are not a part of the 1-core. 

We perform this analysis to identify the "inner sanctum" of a network — the highly resilient and densely interconnected core that often drives systemic behavior, such as spreading information or maintaining market trust. In the context of the Bitcoin OTC dataset, the $k$-core helps distinguish between "tourist" traders and the established "reputation hubs" that anchor the marketplace.

When applying this to a complex dataset like Bitcoin OTC, several analytical considerations arise:
- **Directedness:** In directed networks, $k$-core is typically computed based on the total degree (in-degree + out-degree) within `NetworkX`, though one can choose to focus strictly on in-cores (reputation) or out-cores (activity) depending on the research goal.
- **Weights and Signs:** Standard $k$-core algorithms treat all edges as binary (existing or non-existing). To account for the **signed** nature of this data, it is often necessary to pre-filter the network (e.g., creating a "Trust Core" by removing negative edges) so the $k$-core reflects a core of mutual trust rather than a core of mutual conflict.
- **NetworkX Implementation:** The `nx.core_number(G)` function is the primary tool for this, returning the "shell" index for every node, while `nx.k_core(G)` extracts the actual subgraph for a specific value of $k$.
- **Computational Efficiency:** For a network of ~6,000 nodes, NetworkX handles $k$-core decomposition very efficiently ($O(m)$ time complexity), making it a much faster diagnostic tool than more intensive metrics like Betweenness Centrality.

---

#### The "Peeling" Process (How it works)

$k$-core isn't a single snapshot; it’s an iterative pruning process often called the "shaving" or "peeling" algorithm.
1. Look for all nodes with a degree less than $k$.
2. Remove them from the graph.
3. Removing those nodes might cause their neighbors' degrees to drop. If a neighbor's degree now falls below $k$, remove them too.
4. Repeat until every remaining node has at least $k$ connections to other nodes still in the group.

> It is important to underdtand that this is not **degree filtering** as it has wildly different outcomes. In degree filtering you may select the nodes with $d=100$. However, with k-core, you are first removing all nodes that do not qualify on the degree constraint. It may be the case that the node with $d=100$ connected to 100 1-edges nodes, meaning that once the nodes are pruned, that seemly large node has essentially disapeared. 

---

#### How to Identify the k-core?

Identifying the $k$-core structure is a process of discovery rather than picking an arbitrary number. While you can choose a specific $k$ to filter the network, understanding the structure involves analyzing the entire range of $k$ values the graph can support.

This means that the `k` can be different for each network. Each graph as a maximum possible value of `k` (known as the degeneracy of the graph) and is an intrinsic topological property.
- A Road Network (like a grid) might have a maximum $k$ of only 3 or 4, because no matter how large the network is, no intersection is connected to 100 others in a dense cluster.
- A Scale-Free Network (like Bitcoin OTC) often has a much higher maximum $k$ (e.g., $k=30$ or $k=50$), because the "super-hubs" create a densely interconnected "inner sanctum".

To identify the structure, you don't start with a desired $k$; instead, you calculate the core number for every node in the network. The core number of a node is the largest value $k$ for which that node is still part of the $k$-core.

The "max $k$" (degeneracy) is the most extreme part of the network, but it is rarely the only "important" part. In fact, focusing only on the maximum $k$ is like looking only at the capital city of a country while ignoring its major industrial hubs. Whether a specific $k$ is "best" depends entirely on what you are trying to find.

The max $k$ represents the "inner sanctum"—the most densely interconnected group possible. However, it is often very small (e.g., in Bitcoin OTC, it might only be 20–50 people). The max k is important if you are looking for the "unstoppable" elite who will never lose connectivity, no matter how many other traders leave the platform. However, if you are studying systemic risk. A $k$-core that contains 500 nodes (even if $k$ is lower) is often more "important" to the overall health of the marketplace than a $k$-core of 20 nodes, because the larger group acts as the primary "engine" for liquidity and trades.

Instead of looking for a single "best" $k$, we look at the **distribution of shells**. This tells you the **depth of the hierarchy**:
- If most nodes are in the early shells, e.g. 1,2,3, and then suddenly fall off the the network is flat. It may has a few hub but no interconnected middle class. 
- If you have many nodes at $k=5, k=10,$ and $k=15$, it indicates a rich organizational structure. This means the network has multiple levels of "reputation" and isn't just a binary "Haves vs. Have-nots".

In many studies (like the Kitsak et al.), the goal is to find the most influential spreaders. Sometimes the nodes in the absolute $max-k$ core are too connected to each other; they just talk in a circle. Often, the most important nodes for spreading information (or trust) are in an intermediate $k$-shell. They are connected enough to be reputable, but they still have "tentacles" reaching out into the lower $k$-shells where the rest of the population lives.

### Largest Connected Component (LCC)

In network science, the LCC represents the "Giant Component"—the primary island of connectivity that allows information, trust, or traffic to flow across the majority of the system. In an undirected graph, the LCC is the largest subgraph where every node is reachable from every other node. However, because Bitcoin OTC is a Directed Network, the definition splits into two distinct categories:
- **Weakly Connected Component (WCC):** This ignores the direction of the arrows. It represents the "physical" reach of the network. If two people are in the same WCC, there is some chain of ratings between them, even if the trust doesn't flow both ways.
- **Strongly Connected Component (SCC):** This follows the arrows. It represents mutual reachability. If User A is in the same SCC as User B, there is a path of trust from A to B and a path from B back to A.

The LCC of a **directed network** is usually measured as the **Weakly Connected Component** for robustness studies. The SCC is often much smaller and can even be non-existent in networks that are strictly hierarchical.

The LCC is highly sensitive to the **sign**. If you include negative edges (distrust), your LCC will be large but structurally "toxic." If you filter for only positive edges ($+1$ to $+10$), the LCC shrinks into a **Functional Trust Web**. This version is probably the most relevant.

In a standard connectivity analysis, **weight** is ignored — an edge either exists or it doesn't. However, you can use weights as a **threshold**. For example, "What is the LCC if we only consider 'High Trust' edges (weight $\ge 5$)?". This reveals the "High-Integrity Backbone" of the market.

### Random and Targeted effects on a Weakly Connected Network

- Build a helper function to return the size of the LCC (weakly connected) based on the given network 
- Build function to iteratively attack the network. Default is random attack but includes a param to switch to targetted. 
- The output is a list of tuples where the first value is the number of nodes removed and the second is the size of the lcc (using the helper)

```python
import networkx as nx
import random
import matplotlib.pyplot as plt

def get_lcc_size(G):
    """Helper to return the size of the Largest Weakly Connected Component."""
    if len(G) == 0:
        return 0
    # Weakly Connected Components for Directed Networks
    lcc = max(nx.weakly_connected_components(G), key=len)
    return len(lcc)

def simulate_attack(G_original_network, attack_type='random', steps=50, total_to_remove=1000):
    # total_to_remove=1000 represents ~16% of the Bitcoin network

    G = G_original_network.copy()
    results = []
    
    # Record initial state (0 nodes removed)
    results.append((0, get_lcc_size(G)))
    
    nodes_per_step = total_to_remove // steps
    
    for i in range(steps):
        if attack_type == 'random':
            nodes_to_remove = random.sample(list(G.nodes()), nodes_per_step)

        else: 
            # Targeted: Always pick the current highest degree nodes
            # recalculate degree list at each step because the 'hubs' change
            degrees = dict(G.degree())
            nodes_to_remove = sorted(degrees, key=degrees.get, reverse=True)[:nodes_per_step] # slice top nodes
        
        G.remove_nodes_from(nodes_to_remove)
        current_removed = (i + 1) * nodes_per_step
        results.append((current_removed, get_lcc_size(G))) # append iteration state
        
    return results

# Execute the simulations
random_results = simulate_attack(G, attack_type='random')
targeted_results = simulate_attack(G, attack_type='targeted')
```

- Plot both lines on the same graph
- X-axis is the number of nodes cumlatively removed
- Y-axis is the size of LCC as a percentage of the original network

```python
plt.figure(figsize=(10, 6))

# Extract X and Y for plotting: unpack list of tuples into list coords
x_rand, y_rand = zip(*random_results)
x_targ, y_targ = zip(*targeted_results)

# Normalize Y-axis to percentage of original LCC size
initial_lcc = y_rand[0] # starting point
y_rand_pct = [y / initial_lcc * 100 for y in y_rand]
y_targ_pct = [y / initial_lcc * 100 for y in y_targ]

plt.plot(x_rand, y_rand_pct, label='Random Failure', color='blue', marker='o')
plt.plot(x_targ, y_targ_pct, label='Targeted Attack (Hubs)', color='red', marker='x')

plt.xlabel("Number of Nodes Removed")
plt.ylabel("LCC Size (% of Original)")
plt.title("Network Robustness: Bitcoin OTC")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

#### Superhub Removal

Here I wanted to explore the individual effect of removing the networks single "super-hub". This is the node that has the highest degree by some margin. The code is set up to remove it in the middle of a random attack and as expected it has a huge impact on the LCC

```python
import pandas as pd
import networkx as nx
import random
import matplotlib.pyplot as plt


def hybrid_attack_simulation(G_orig, total_nodes_to_remove=1000, steps=50):
    G = G_orig.copy()
    
    # Identify the specific Super Hub (Node with max degree)
    super_hub = max(dict(G.degree()).items(), key=lambda x: x[1])[0]
    print(f"Target identified: Super Hub ID {super_hub}")
    
    results = []
    nodes_per_step = total_nodes_to_remove // steps
    mid_step = steps // 2
    
    # Record initial state
    initial_lcc = len(max(nx.weakly_connected_components(G), key=len))
    results.append((0, initial_lcc))
    
    for i in range(1, steps + 1):
        if i == mid_step:
            # SHOCK STEP: Remove the Super Hub specifically
            # Plus enough random nodes to match the nodes_per_step
            other_nodes = [n for n in G.nodes() if n != super_hub]
            to_remove = random.sample(other_nodes, nodes_per_step - 1)
            to_remove.append(super_hub)
            print(f"Step {i}: SHOCK - Removing Super Hub {super_hub}")
        else:
            # NORMAL STEP: Pure Random Removal
            # Ensure we don't accidentally pick the super hub early
            available_nodes = [n for n in G.nodes() if n != super_hub]
            to_remove = random.sample(available_nodes, nodes_per_step)
            
        G.remove_nodes_from(to_remove)
        
        # Track LCC
        current_lcc = len(max(nx.weakly_connected_components(G), key=len))
        results.append((i * nodes_per_step, current_lcc))
        
    return results, (mid_step * nodes_per_step)

# Run Simulation
results, shock_point = hybrid_attack_simulation(G)

# Plotting the Shock
x, y = zip(*results)
y_pct = [val/y[0]*100 for val in y]

plt.figure(figsize=(10, 6))
plt.plot(x, y_pct, marker='o', linestyle='-', color='blue', label='Random Attack + Mid-point Shock')
plt.axvline(x=shock_point, color='red', linestyle='--', label='Super Hub Removed')

plt.annotate('Structural Shock', xy=(shock_point, y_pct[len(y_pct)//2]), 
             xytext=(shock_point+100, y_pct[len(y_pct)//2]+10),
             arrowprops=dict(facecolor='black', shrink=0.05))

plt.xlabel("Nodes Removed")
plt.ylabel("LCC Size (% of Original)")
plt.title("Vulnerability to Single Super Hub Removal")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

#### Building Trust Networks: Positive or Highly Positive

- This function allows the original network to be filtered based on the sign and weight

```python
import networkx as nx

def get_functional_trust_network(G_orig, min_weight=1):
    """
    Filters the network to include only trusted edges.
    
    Parameters:
    - G_orig: The original Bitcoin OTC DiGraph
    - min_weight: The threshold for trust (default 1, meaning all positive ratings).
                  Set higher (e.g., 5 or 8) for a 'high-integrity' network.
    """
    # Create a list of edges that meet the criteria
    # Bitcoin OTC ratings are from -10 to +10. 
    # Positive trust starts at 1.
    trusted_edges = [
        (u, v, d) for u, v, d in G_orig.edges(data=True) 
        if d['weight'] >= min_weight
    ]
    
    # Initialize a new Directed Graph
    G_trust = nx.DiGraph()
    
    # We add all nodes from the original graph to maintain the same node ID space, 
    # though many may become isolated (degree 0).
    G_trust.add_nodes_from(G_orig.nodes())
    G_trust.add_edges_from(trusted_edges)
    
    # Remove isolated nodes if you want a cleaner view for visualization
    G_trust.remove_nodes_from(list(nx.isolates(G_trust)))
    
    return G_trust

# Example usage:
# Baseline Trust Network (All positive ratings)
G_pos = get_functional_trust_network(G, min_weight=1)

# Highly Trustful Network (Only ratings of +8 and above)
G_elite_trust = get_functional_trust_network(G, min_weight=5)
```

```python
# 1. Handle Positive Trust Network
wcc_pos = list(nx.weakly_connected_components(G_pos))
lcc_nodes_pos = max(wcc_pos, key=len)

# 2. Handle Highly Positive Trust Network (Recalculate here!)
wcc_elite = list(nx.weakly_connected_components(G_elite_trust))
lcc_nodes_elite = max(wcc_elite, key=len)

# Print Results using the distinct LCC variables
print(f"Total Nodes in System: {len(G_pos)}")

print(f"Positive Trust LCC Size: {len(lcc_nodes_pos)}")
print(f"Highly Positive Trust LCC Size: {len(lcc_nodes_elite)}")

# Use the specific numerator for each category
print(f"Positive Trust LCC Percentage: {len(lcc_nodes_pos) / len(G_pos) * 100:.2f}%")
print(f"Highly Positive Trust LCC Percentage: {len(lcc_nodes_elite) / len(G_elite_trust) * 100:.2f}%")
```

```python
import networkx as nx
import random
import matplotlib.pyplot as plt

def get_lcc_size(network, graph_type="directed"):
    """Helper to return the size of the Largest Weakly Connected Component."""
    if len(network) == 0:
        return 0
    
    if graph_type == "directed":
        # Weakly Connected Components for Directed Networks
        lcc = max(nx.weakly_connected_components(network), key=len)
    elif graph_type == "undirected": 
        lcc = max(nx.connected_components(network), key=len)

    return len(lcc)

def simulate_attack(G_original_network, attack_type='random', steps=50, total_to_remove=1000, graph_type="directed"):
    # total_to_remove=1000 represents ~16% of the Bitcoin network

    G = G_original_network.copy()
    results = []
    
    # Record initial state (0 nodes removed)
    results.append((0, get_lcc_size(G, graph_type=graph_type)))
    
    nodes_per_step = total_to_remove // steps
    
    for i in range(steps):
        if attack_type == 'random':
            nodes_to_remove = random.sample(list(G.nodes()), nodes_per_step)

        else: 
            # Targeted: Always pick the current highest degree nodes
            # recalculate degree list at each step because the 'hubs' change
            degrees = dict(G.degree())
            nodes_to_remove = sorted(degrees, key=degrees.get, reverse=True)[:nodes_per_step] # slice top nodes
        
        G.remove_nodes_from(nodes_to_remove)
        current_removed = (i + 1) * nodes_per_step
        results.append((current_removed, get_lcc_size(G, graph_type=graph_type))) # append iteration state
        
    return results
```

## Grid: [US power grid](http://konect.cc/networks/opsahl-powergrid/)

The opsahl-powergrid network is a seminal dataset in network science. This undirected network contains information about the power grid of the Western States of the United States of America. An edge represents a power supply line. A node is either a generator, a transformator or a substation. Its structure is shaped by geographic reality and engineering necessity rather than social growth.

> **Source:** Watts, D. J., Strogatz, S. H., 1998. [Collective dynamics of “small-world” networks](https://toreopsahl.com/datasets/#uspowergrid). Nature 393, 440-442.

### Dataset Overview
- **Nodes:** 4,941 (generator, transformator or substation)
- **Edges:** 6,594 (Supply Lines)
- **Network Type:** Undirected, Unweighted
- **Domain:** Infastructure

---

### Key Structural Features:
##### Degree-Bounded Topology (Non-Scale-Free):
The power grid does not follow a power law and is not scale-free because physical substations have a finite limit on how many high-voltage lines they can physically accommodate. Most nodes have a very low degree (usually between 2 and 5), and there are no "super-hubs" with thousands of connections like those found in the Bitcoin network. On a log-log plot, the degree distribution will show a steep exponential decay rather than a long, straight-line tail, meaning the network has a "typical" scale of connectivity.

##### High Local Clustering and Small-Worldness:
The network exhibits a high clustering coefficient, which is a hallmark of "Small-World" networks. This means that if a substation is connected to two others, there is a very high probability those two are also connected to each other, forming local triangles. These triangles provide local redundancy, ensuring that if a single line fails, power can often be rerouted within the immediate neighborhood. Despite being constrained by geography, the network still manages to have a relatively low "average path length" due to a few long-range lines that act as shortcuts across the map.

##### Decentralized Robustness and Hub Insignificance:
Because the power grid lacks massive central hubs, it behaves very differently under attack than a scale-free system. In your Bitcoin experiment, removing the top hubs shattered the network; in the power grid, removing the "highest degree" nodes (which might only have 10–20 connections) will have a minimal impact on the overall Largest Connected Component (LCC). The network is remarkably resilient to targeted attacks because its connectivity is distributed across a mesh-like structure rather than being anchored to a few specific individuals.

##### Vulnerability to Random Failure and Cascading Effects:
While it resists targeted attacks, the grid is surprisingly sensitive to random failures because it lacks the "hyper-connectivity" that hubs provide. Since the network is relatively sparse and spread out, losing a few random nodes can easily isolate "islands" of the grid, especially in rural areas where connectivity is lower. In real-world terms, this structure is susceptible to cascading failures, where the loss of one line puts too much strain on its neighbors, potentially leading to a widespread blackout even without a coordinated attack on a hub.
