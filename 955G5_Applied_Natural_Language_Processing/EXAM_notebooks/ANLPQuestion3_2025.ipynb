a)

i) Explain what it meant by a bag-of-words representation of a document.

[2 marks]

A BoW represention is a data structure where the features (word) are recorded by have no order or sequence. It comes in a [key:value] form whereby the key is the feature and the value can either be a boolean or a frequence count. If you were to construct a BoW from a document, the keys of corresponding BoW would tell you the Vocabulary and the value would either confirm that is a feature of the document, or that it is a feature and how many times it occured.

ii) Construct a bag-of-words representation of each of the sentences in `mandela` and store it in `mreps`.  You do not need to carry out any other pre-processing of the sentences.

[3 marks]

```python
from nltk import FreqDist

mreps = []

for sent in mandela:
    mreps.append(FreqDist(sent))

mreps
```

b)

i) For each token in the vocabulary of the sentences stored in `mandela`, calculate its document frequency, i.e., the number of sentences that it occurs in, and store this in `docfreq`

[5 marks]

```python
docfreq = {}

for token in set([token for sent in mandela for token in sent]):
    for BoW in mreps:
        current = docfreq.get(token,0)
        if BoW.get(token,0) > 0:
            current += 1
        docfreq[token] = current

print(docfreq[","])
docfreq
```

ii) Which token occurs in more sentences in the sample than any other token?  And in how many sentences does that token occur?

[5 marks]

```python
most_freq= max(docfreq, key=docfreq.get)
count = docfreq[most_freq]

print(f"The most frequent token is '{most_freq}'")
print(f"It occurs {count} times")
```

iii) Write python code to calculate the tfidf values associated with each of the tokens in the first sentence in the sample?

[10 marks]

```python
import math

# WHERE docfreq = (df) and mreps = (tf)

first_sent = mandela[0]
first_bow = mreps[0]
n_docs = len(mandela)

idf = {}

for word in first_sent:
    idf[word] = math.log(n_docs/docfreq[word])

tf_idf = {}

for token in first_sent:
    tf_idf[token] = first_bow[token] * idf[token]

print(tf_idf)
```

ii) What is meant if we say that a word is ambiguous with respect to part-of-speech?  Give an example of a word in `tagged_sentences[2]` which could be said to be ambiguous with respect to part-of-speech.  What information does a HMM part-of-speech tagger use to try to resolve the ambiguity?

[7 marks]

A word is ambigious if in isolation, and without surrounding or domain context, the word could be understood and used in more than 1 different way. This can mean that the word has an entirely different sense and it could also mean that the POS is it a part of is different. An example from the text is "will" which has been tagged is "MD" which is a Modal Verb. A modal is an auxiliary verb that expresses necessity or possibility. This is probably the correct tag for this sentence but in isolation "will" could be a Noun (NN) refereing to the legal document or a persons name (NNP). Both of these noun examples would be capitalized but as our text has been pre-case normalised wouldn't be able to tell the different. A HMM implements two assumptions: vertical and horizontal. The horizontal assumption says that the current words tag depends only on the tag of the previous word. Therefore, in tagging, the HMM ties in the information from the previous word though its tag. If the previous tag was a Personal Pronoun such as "I" then we would know that it is not "will" as in legal document, nor "will" the person as this would be "I, Will". Though the latter also depends on the genre/domain as even though our text includes punctuation, if it comes from a source where punctuation is always correct or possible (speech-to-text), then it still could be that. 

iii) Find the token in the sample with the highest part-of-speech tag ambiguity in terms of the entropy of its tag distribution.  Explain how you arrived at your answer.

Comment on the tags assigned to this token by the pos tagger.

[15 marks]

```python
tags = {}

for sent in tagged_sentences:
    for token,tag in sent:
        outer = tags.get(token,{})
        count = outer.get(tag,0)+1
        outer[tag] = count
        tags[token] = outer

for key in tags.keys():
    denom = sum(tags[key].values())
    current = tags.get(key,{})
    for tag in current:
        current[tag] = current[tag]/denom

tags

ents = {}

for token,probs in tags.items():
    ent = 0
    for prob in probs.values():
        ent -= prob * math.log2(prob)
    ents[token] = ent

highest_ent_token= max(ents, key=ents.get)
highest_ent_count = ents[highest_ent_token]

print(f"The token with the highest ent is '{highest_ent_token}'")
print(f"It has an entropy of {highest_ent_count}")

print(f"It has the tags and distributions are: {tags[highest_ent_token]}")
```
