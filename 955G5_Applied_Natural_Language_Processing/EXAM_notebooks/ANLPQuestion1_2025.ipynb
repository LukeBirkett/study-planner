iv) Calculate and print the number of hapax legomena in `x`

[3 marks]

```python
hl = {}
for token in doc_list:
    hl[token] = hl.get(token, 0)+1

hapaxes = [word for word in hl if hl[word] == 1]
print(len(hapaxes))
```
