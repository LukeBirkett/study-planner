Here is a basic function for implementing the naive method, assuming square matrices are input:

```python
def naivemult(A,B):

    (m,n) = np.shape(A)
        
    C = np.zeros([m,m])
    
    for i in range(m):
        for j in range(m):
            for k in range(m):
                C[i,j] += A[i,k] * B[k,j]
        
    return C
```

Let's import the timeit function from Lab 2.

```python
import time

def timeit(somefunc,*args,repeats=10,**kwargs):
    times=[]
  
    while repeats>0:
        starttime=time.time()
        ans=somefunc(*args,**kwargs)
        endtime=time.time()
        timetaken=endtime-starttime
        times.append(timetaken)
        repeats-=1
    
    mean=np.mean(times)
    stdev=np.std(times)
 
    return (mean,stdev)
```

And the makerandlist function as well.

```python
import random

def makerandlist(n,startno=1,endno=100):
    result=[]
    while n >0:
        result.append(random.randint(startno,endno))
        n-=1
    return result
```

For a range of values of n, lets generate random nxn matrices, and time how long it takes to multiply them. We store the values of n in one list, xs, and the run-times in another, ys.

```python
xs=[]
ys=[]
for n in range(20,120,5):
    xs.append(n)
    A=np.array([makerandlist(n) for i in range(n)])
    B=np.array([makerandlist(n) for i in range(n)])
    (mean,stdev)=timeit(naivemult,A,B,repeats=10)
    ys.append(mean)
```

Then we make the plot

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.scatter(xs,ys)
plt.xlabel("Dimension of matrix")
plt.ylabel("Average time to multiply")
plt.show()
```

Take logarithms:

```python
logxs=[np.log10(x) for x in xs]
logys=[np.log10(y) for y in ys]

plt.scatter(logxs,logys)
plt.xlabel("Log of dimension of matrix")
plt.ylabel("Log of average time to multiply")
plt.show()
```

By inspection, we can see that as the log of the dimension increases from 1.5 to 2, the logarithm of the run-time increases from approximately -2 to approximately -0.5. Thus the gradient on this plot is approximately 3, suggestive of $O(n^3)$ time complexity.

Note I've done this calculation based on the larger sizes of matrix - you get a better approximation to the time complexity by considering larger data, since time complexity captures the asymptotic behaviour of the run-time.

Here is code for implementing Strassen's method:

```python
def segment(M):
    (mm,nn)=np.shape(M)
    n = int(mm/2)
    
    a=M[:n,:n]
    b=M[:n,n:]
    c=M[n:,:n]
    d=M[n:,n:]
    
    return (a,b,c,d)


def build(a,b,c,d):
    
    (mm,nn)=np.shape(a)
    C=np.zeros([2*mm,2*mm])
    
    C[:mm,:mm]=a
    C[:mm,mm:]=b
    C[mm:,:mm]=c
    C[mm:,mm:]=d
            
    return C
    

def strassen(M1, M2):
    
    if len(M1) <= 2:
        return naivemult(M1,M2)
    
    (a,b,c,d) = segment(M1)

    #note: e g f h NOT e f g h 
    (e,g,f,h) = segment(M2)

    P1 = strassen(a, g-h)
    P2 = strassen(a+b, h)
    P3 = strassen(c+d,e)
    P4 = strassen(d,f-e)
    P5 = strassen(a+d,e+h)
    P6 = strassen(b-d,f+h)
    P7 = strassen(a-c,e+g)
    
    r = P5+P4-P2+P6
    s = P1+P2
    t = P3+P4
    u = P5-P3-P7+P1
    
    C = build(r,s,t,u)
    
    return C
```
