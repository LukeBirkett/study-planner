```python
def predict_capital(trainingpair, countries):
    capitals = [mymodel.most_similar(positive=[trainingpair[0],country],negative=[trainingpair[1]])[0][0] for country in countries]
    return capitals                                 
        
```

### Exercise 2

```python
import numpy as np
relation='capital-common-countries'

allpairs=testtuples[relation]

def evaluate_capitals():
    random.shuffle(allpairs)
    capitals,countries=zip(*allpairs)
    accuracies=[]
    for tp in allpairs:
        print("Training on {}".format(tp))
        predictions=predict_capital(tp,countries)
        correct=0
        for country, pre, cap in zip(countries, predictions, capitals):
            #print("{}, {}, {}".format(country, pre, cap))
            if pre==cap and tp[0]!=cap:
                #ignore prediction for training pair and check prediction otherwise
                correct+=1
        accuracy=correct/(len(predictions)-1)
        print("Accuracy is {}".format(accuracy))
        accuracies.append(accuracy)
    print("Mean accuracy: {}, sd: {}".format(np.mean(accuracies),np.std(accuracies)))

evaluate_capitals()
```

### Exercise 3
Accuracy and computation time will be improved if you make it a multiple choice test i.e., which is the most similar out of a selection of 6 possible answers (as in the original paper).  You could use all of the known capitals as foils.

You should try out this variant and see the effect on performance.


```python
import operator,math

#for some reason np.dot(a,a) was giving strange (zero) result
def dot(a,b):
    return sum(a*b)

def cosine(a,b):
    #print(type(a),a)
    num=dot(a,b)
    adota=dot(a,a)
    bdotb=dot(b,b)
    #print(num,adota,bdotb)
    if adota == 0 or bdotb==0 or num ==0:
        
        return 0
    else:
        return num/math.sqrt(adota*bdotb)
            
      
def predict_capital_multiplechoice(trainingpair, countries,choices):
    capitals=[]
    for country in countries:
        offset=mymodel[trainingpair[0]]+mymodel[country]-mymodel[trainingpair[1]]
        
        #print(type(offset),offset)
        sims=[(c,cosine(offset,mymodel[c])) for c in choices]
        sortedlist=sorted(sims,key=operator.itemgetter(1),reverse=True)
    
        #print(sortedlist)
        capitals.append(sortedlist[0][0])
    return capitals  
```

```python
def evaluate_capitals():
    random.shuffle(allpairs)
    capitals,countries=zip(*allpairs)
    accuracies=[]
    for tp in allpairs:
        print("Training on {}".format(tp))
        predictions=predict_capital_multiplechoice(tp,countries,capitals)
        correct=0
        for country, pre, cap in zip(countries, predictions, capitals):
            print("{}, {}, {}".format(country, pre, cap))
            if pre==cap:
                correct+=1
        accuracy=correct/len(predictions)
        print("Accuracy is {}".format(accuracy))
        accuracies.append(accuracy)
    print("Mean accuracy: {}, sd: {}".format(np.mean(accuracies),np.std(accuracies)))

evaluate_capitals()
```
